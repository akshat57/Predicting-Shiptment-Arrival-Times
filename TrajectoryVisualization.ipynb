```python
import pandas as pd
import numpy as np
import folium
import io
from PIL import Image
```

```python
train_data = pd.read_pickle('../Processed_Data/train_set.pkl')
test_data = pd.read_pickle('../Processed_Data/test_set.pkl')
```

```python
folium_map = folium.Map( location=[40.58338,-73.85482])
```

```python
trajectory = pd.concat([pd.DataFrame(np.array(train_data[30]).T, columns=['Lat','Long','Time'])],ignore_index=True)    
```

```python
trajectory.apply(lambda row:folium.CircleMarker(location=[row["Lat"], row["Long"]] ).add_to(folium_map), axis=1)
```
